<b><i>Add and remove functions</i></b></div>

```python
carCompanies = ["Maruti Suzuki",
"Hyundai",
"Tata Motors",
"Mahindra & Mahindra",
"Kia",
"Toyota",
"Honda",
"Renault"]
print(carCompanies)
```

```python
# Use of append function to add new element at the end
carCompanies.append("BMW")
print(carCompanies)
```

```python
# Removing the value from the list using remove function
carCompanies.remove("Kia")
print(carCompanies)
```

```python
# Removing the last value from the list using pop function
carCompanies.pop()
print(carCompanies)
```

```python
# Removing an item at the specific position using pop function
carCompanies.pop(3)
print(carCompanies)
```

```python
# Inserting an item at a specific index position
carCompanies.insert(3,"Kia")
print(carCompanies)
```

```python
# Clearing list
carCompanies.clear()
print(carCompanies)
```

```python
# Deleting a list
del carCompanies
```

<b><i>Joining and Extending a list</i></b></div>

```python
# List 1
IndianBikes = ["Hero", "Bajaj", "Jawa", "TVS", "Royal enfield"]
# List 2
ForeignBikes = ["Honda", "KTM", "Harley davidson", "Yamaha", "Suzuki"]
```

```python
# Joining list 1 and list 2
allBikes = IndianBikes + ForeignBikes
print(allBikes)
```

```python
# Extending list 1 with list 2
# list 1 is updated based on list 2 and it will not create any new variable
IndianBikes.extend(ForeignBikes)
print(IndianBikes)
```

```python
# Use of append with list 1 and list 2
# List 1
IndianBikes = ["Hero", "Bajaj", "Jawa", "TVS", "Royal enfield"]
# List 2
ForeignBikes = ["Honda", "KTM", "Harley davidson", "Yamaha", "Suzuki"]
IndianBikes.append(ForeignBikes)
print(IndianBikes)
```

append will nest list inside the list</div>

<b><i>Sorting and Reversing a List</i></b>
</div>

```python
# Student list
MscGIS = ["Sumanth", "Inchara", "Adithia", "Yashaswini", "Nithya","Manoj", "Divya", "Shivaprasad"]
```

```python
# Ascending sort
MscGIS.sort()
print(MscGIS)
```

**|Method| |Description|**
- append()=Adds an element at the end of the list
- clear()=Removes all the elements from the list
- copy()=Returns a copy of the list
- count()=Returns the number of elements with the specified value
- extend()=Add the elements of a list (or any iterable), to the end of the current list
- index()=Returns the index of the first element with the specified value
- insert()=Adds an element at the specified position
- pop()=Removes the element at the specified position
- remove()=Removes the item with the specified value
- reverse()=Reverses the order of the list
- sort()=Sorts the list
