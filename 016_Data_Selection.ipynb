```python
# Creating deep copy of the data.
dfNew = df.copy(deep = True)
```

```python
dfNew
```

```python
col = ['Name','Level']
dfNew = dfNew[col]
```

```python
dfNew
```

```python
# Sorting the values of column
dfNew.sort_values('Name',ascending = True)
```

```python
# Use of inplace function to store the sorted value within the same variable
dfNew.sort_values('Name',ascending = False, inplace = True)
```
