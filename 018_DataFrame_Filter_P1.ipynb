```python
# Importing pandas
import numpy as np
import pandas as pd
```

```python
# Reading csv file
df = pd.read_csv('EmpDataCleaned.csv')
```

```python
df
```

```python
# deleting the unwanted columns
del df['Unnamed: 0']
```

```python
df
```

```python
# Retrive rows by index position with iloc[] function (index locate)
df.iloc[0]
```

```python
# Retrive rows by index position with iloc[] function (index locate)
df.iloc[10]
```

```python
# Retrive rows by index position with iloc[] function (index locate)
df.iloc[15]
```

```python
# Locating multiple rows
df.iloc[[0,5,10,15]]
```

```python
# Locating series of rows inbetween selected row number
df.iloc[5:11]
```

```python
# Accessing the element based on row and column intersection
# if we want to access 8th row(PAVAN KUMAR G) and his working team index position is 3
df.iloc[8,3]
```

```python
# if we want to access 6th row(NIRANJAN) and his height index position is 6
# Niranjan height is 
df.iloc[6,6]
```

```python
# Working with loc[] accessor
# To work with loc[] your index should be in index label
df
```

```python
# Setting an index label
df.set_index(keys = 'Name', inplace = True)
```

```python
df
```

```python
# After setting index label you should sort your index label
# Because sorted collection speed is greater than unsorted collection speed
df.sort_index(inplace = True)
```

```python
df
```

```python
# Extracting rows using loc[] and index labels
df.loc['YASHASWINI G K']
```

```python
# Extracting rows using loc[] and index labels
df.loc['SANJOK CHAWHAN']
```

```python
# Locating multiple rows
required = ['YASHASWINI G K','SANJOK CHAWHAN']
df.loc[required]
```

```python
# Locating multiple rows
df.loc[['YASHASWINI G K','SANJOK CHAWHAN']]
```

```python
# filtering rows starting from A to M use next letter [A:N]
df.loc['A':'N']
```

```python
# filtering rows starting from N to S use next letter [N:T]
df.loc['N':'T']
```

```python
# Accessing the element based on row and column intersection
df.loc['SHIVAPRASAD N','Team']
```

```python
# Saving Extracted data based on loc and iloc
df.loc['N':'T']
```

```python
# Storing the extracted item to a new variable
Emp_N2S = df.loc['N':'T']
```

```python
Emp_N2S
```

```python
# Saving the cleaned data
Emp_N2S.to_csv('selected_EMP.csv')
```
