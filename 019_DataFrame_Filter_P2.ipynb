```python
# Importing pandas and numpy
import numpy as np
import pandas as pd
```

```python
# Reading the data
df = pd.read_csv('EmpDataCleaned.csv')
```

```python
df
```

```python
# Deleting unwanted columns
del df['Unnamed: 0']
```

```python
df
```

```python
# Checking how many males
df['Gender'].value_counts()
```

```python
# Generaing a seperate frame for males
df['Gender'] == 'Male'
# This code checks the condition and outputs in booleans
```

```python
# Generating a subset of Males from main dataframe
df[df['Gender'] == 'Male']
```

```python
# Generating Female data from main data frame
df[df['Gender']=='Female']
```

```python
# Generating HR frame
df[df['Team']=='HR']
```

```python
# Checking cardinality
df['Team'].unique()
```

```python
# Generate Accounts frame
df[df['Team']=='Accounts']
```

```python
# Generating Accounts and HR team together
condition1 = df['Team'] == 'HR'
condition2 = df['Team'] == 'Accounts'
```

```python
df[(condition1) | (condition2)]
```

```python
# Generating male and HR
condition1 = df['Gender'] == 'Male'
condition2 = df['Team'] == 'HR'
```

```python
df[(condition1) & (condition2)]
```

```python
# Age >= 23
df[df['Age']>=23]
```

```python
# Those who having loan
df[df['Loans']=='Yes']
```

```python
# Height between 160 to 170 
df[(df['Height'] >= 160) & (df['Height']<=170)]
```

```python
# To save employee with loan
EmpLoan = df[df['Loans']=='Yes']
```
