```python
# Classifying the column values of BMI
def classifyBMI(bmi):
    if bmi < 18.49:
        return 'Under Weight'
    elif bmi >= 18.49 and bmi < 24.99:
        return 'Normal Weight'
    elif bmi >= 24.99 and bmi <29.99:
        return 'Overweight'
    else:
        return 'Obese'
```

```python
df['BMI_Result'] = df['BMI'].apply(classifyBMI)
```
