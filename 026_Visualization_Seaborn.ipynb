```python
# Importing matplotlib and seaborn
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Compressing the graph within the jupyter
%matplotlib inline
```

```python
# loading default data
df = sns.load_dataset('tips')
```

```python
df
```

```python
# Plotting total number of male and female
plt.figure(figsize=(5,3))
sns.countplot(x='sex', data=df)
plt.show()
```

```python
# plotting conunt plot for day
plt.figure(figsize=(10,3))
sns.countplot(x='day', data=df)
plt.show()
```

```python
# Generating line graph 
plt.figure(figsize=(10,3))
sns.lineplot(x='day',y='total_bill',data=df)
plt.show()
```

```python
# Changing x values to vertical
plt.figure(figsize=(10,3))
sns.lineplot(x='day',y='total_bill',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
# Generating barplot
plt.figure(figsize=(10,3))
sns.barplot(x='day',y='total_bill',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
# Generating scatter plot
plt.figure(figsize=(10,3))
sns.scatterplot(x='tip',y='total_bill',data=df)
plt.show()
```

```python
# Segregating male and female
plt.figure(figsize=(5,3))
sns.scatterplot(x='tip',y='total_bill',data=df, hue='sex')
plt.show()
```
