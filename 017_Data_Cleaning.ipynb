```python
# Importing pandas
import numpy as np
import pandas as pd
```

```python
# Reading csv file
df = pd.read_csv("empData.csv")
```

```python
df
```

```python
# Renaming columns
df.columns
```

```python
# Renaming column
newColName = ['Name','Salary','Gender','Team','Status','Age','Height','Peformance','Loans','LOP']
df.columns = newColName
```

```python
df
```

```python
# Checking completenes of the data
# Checking null values
df.isnull()
```

```python
df.isnull().sum()
```

```python
# Deleting the column
df.drop('Peformance', axis = 1, inplace = True)
```

```python
df
```

```python
# Deleting columns
del df['LOP']
```

```python
df
```

```python
# Deleting the rows
# use drop function
df.drop([1], axis = 0, inplace = True)
```

```python
df
```

```python
# Reseting index after deleting any rows
df.reset_index(inplace = True)
```

```python
df
```

```python
# Deleting columns
del df['index']
```

```python
df
```

```python
# Checking the duplicated rows
df.duplicated()
```

```python
# Checking total number of duplicates
df.duplicated().sum()
```

```python
# if there is a duplicate then 
df.drop_duplicates()
```

```python
# Checking unique counts
df['Gender'].unique()
```

```python
df['Team'].unique()
```

```python
df['Status'].unique()
```

```python
df['Loans'].unique()
```

```python
# Counting the unique item
df['Gender'].nunique()
```

```python
# Counting category wise total
df['Gender'].value_counts()
```

```python
# Content formatting
# ex: df.replace({0: 10, 1: 100})
df['Gender'].replace({'m':'Male', 'f':'Female'}, inplace = True)
```

```python
# Replacing values in a row
df['Loans'].replace({'y':'Yes', 'n':'No'}, inplace = True)
```

```python
# Replacing values in a row
df['Status'].replace({'p':'Permanent', 't':'Temporary'}, inplace = True)
```

```python
# Checking unique values
df['Team'].unique()
```

```python
# Replacing values in a row
replace = {'hr':'HR',
           'fin':'Finance',
           'rd':'R&D',
           'bo':'Back Office',
           'acc':'Accounts',
          'bd': 'Business Development'}
df['Team'].replace(replace, inplace = True)
```

```python
df
```

```python
# Saving the cleaned data
df.to_csv('EmpDataCleaned.csv')
```
