```python
# Import pandas and numpy
import numpy as np
import pandas as pd
```

```python
# Reading csv file
df = pd.read_csv('EmpDataCleaned.csv')
```

```python
df
```

```python
# Using groupby() function to group dataframe to unique values of a column
group = df.groupby('Gender')
```

```python
# Calling group with Gender Male
group.get_group('Male')
```

```python
# Calling group with Gender Female
group.get_group('Female')
```

```python
# Calculating mean height of female group
group.get_group('Female')['Height'].mean()
```

```python
# Checking minimum height of female group
group.get_group('Female')['Height'].min()
```

```python
# Checking maximum height of female group
group.get_group('Female')['Height'].max()
```

```python
# Calculating sum of salary of female group
group.get_group('Female')['Salary'].sum()
```

```python
# Calculating sum of salary of male group
group.get_group('Male')['Salary'].sum()
```

```python
# Groupby methods
len(group)
```

```python
# Checkinh group size
group.size()
```

```python
# Checkinh First value in group
group.first()
```

```python
# Checkinh least value in a group
group.last()
```

```python
# Calling index values of group in dictionary
group.groups
```

```python
# Aggregate Function
group.agg({'Age': 'mean', 'Salary':'sum'})
```

```python
# Aggregate Function
group.agg({'Age': 'mean', 'Salary':'mean'})
```
