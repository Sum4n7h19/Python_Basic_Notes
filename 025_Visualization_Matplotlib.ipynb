```python
# Generating multiple y values
age = [23,24,25,26,27,28,29]
height = [5.6,5.8,5.9,6.3,5.2,5.9,5.8]
weight = [65,78,80,56,69,70,85]
```

```python
plt.figure(figsize=(5,3))
plt.plot(age,height)
plt.plot(age,weight)
plt.show()
```

```python
# Adding label to the graph
plt.figure(figsize=(5,3))
plt.plot(age,height,label='height')
plt.plot(age,weight,label='weight')
plt.legend()
plt.show()
```

```python
# Adding axis name
plt.figure(figsize=(5,3))
plt.plot(age,height,label='height')
plt.plot(age,weight,label='weight')
plt.xlabel('age')
plt.ylabel('height and weight')
plt.legend()
plt.show()
```

```python
# Customizing font size of axis labels
plt.figure(figsize=(5,3))
plt.plot(age,height,label='height')
plt.plot(age,weight,label='weight')
plt.xlabel('age',fontsize=12)
plt.ylabel('height and weight',fontsize=12)
plt.legend()
plt.show()
```

```python
# Adding title to the graph
plt.figure(figsize=(5,3))
plt.plot(age,height,label='height')
plt.plot(age,weight,label='weight')
plt.xlabel('age',fontsize=12)
plt.ylabel('height and weight',fontsize=12)
plt.title('Age vs Height and Weight',fontsize = 12)
plt.legend()
plt.show()
```

```python
# Stylizing a line
# Line styles (dashed, dashdot,solid,dotted,--,-.,-,:)
plt.figure(figsize=(5,3))
plt.plot(age,height,label='height',linestyle='dashed')
plt.plot(age,weight,label='weight',linestyle='dotted')
plt.xlabel('age',fontsize=12)
plt.ylabel('height and weight',fontsize=12)
plt.title('Age vs Height and Weight',fontsize = 12)
plt.legend()
plt.show()
```

```python
# placing legend outside the graph
plt.figure(figsize=(5,3))
plt.plot(age,height,label='height',linestyle='dashed')
plt.plot(age,weight,label='weight',linestyle='dotted')
plt.xlabel('age',fontsize=12)
plt.ylabel('height and weight',fontsize=12)
plt.title('Age vs Height and Weight',fontsize = 12)
plt.legend(loc='upper right',bbox_to_anchor=(1.3,1))
plt.show()
```

```python
# Changing line width
plt.figure(figsize=(5,3))
plt.plot(age,height,label='height',linestyle='dashed',linewidth=3)
plt.plot(age,weight,label='weight',linestyle='dotted',linewidth=3)
plt.xlabel('age',fontsize=12)
plt.ylabel('height and weight',fontsize=12)
plt.title('Age vs Height and Weight',fontsize = 12)
plt.legend(loc='upper right',bbox_to_anchor=(1.3,1))
plt.show()
```

```python
# Adding grids to the graph
plt.figure(figsize=(5,3))
plt.plot(age,height,label='height',linestyle='dashed',linewidth=3)
plt.plot(age,weight,label='weight',linestyle='dotted',linewidth=3)
plt.xlabel('age',fontsize=12)
plt.ylabel('height and weight',fontsize=12)
plt.title('Age vs Height and Weight',fontsize = 12)
plt.legend(loc='upper right',bbox_to_anchor=(1.3,1))
plt.grid()
plt.show()
```

```python
# Adding color to the grids
plt.figure(figsize=(5,3))
plt.plot(age,height,label='height',linestyle='dashed',linewidth=3)
plt.plot(age,weight,label='weight',linestyle='dotted',linewidth=3)
plt.xlabel('age',fontsize=12)
plt.ylabel('height and weight',fontsize=12)
plt.title('Age vs Height and Weight',fontsize = 12)
plt.legend(loc='upper right',bbox_to_anchor=(1.3,1))
plt.grid()
plt.show()
```

```python
# Adding color to the grids
plt.figure(figsize=(5,3))
ax = plt.axes()
ax.set_facecolor('lightgrey')
plt.plot(age,height,label='height',linestyle='dashed',linewidth=3)
plt.plot(age,weight,label='weight',linestyle='dotted',linewidth=3)
plt.xlabel('age',fontsize=12)
plt.ylabel('height and weight',fontsize=12)
plt.title('Age vs Height and Weight',fontsize = 12)
plt.legend(loc='upper right',bbox_to_anchor=(1.3,1))
plt.grid()
plt.show()
```

```python
# Generating subplots horizontally
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.plot(age,height)
plt.subplot(1,2,2)
plt.plot(age,weight)
plt.show()
```

```python
# Generating subplot vertically
plt.figure(figsize=(5,10))
plt.subplot(2,1,1)
plt.plot(age,height)
plt.subplot(2,1,2)
plt.plot(age,weight)
plt.show()
```

```python
# Bar Graph
name = ['Sumanth','Inchara','Manoj','Nithya','Sanjay']
score = [8.4,8.9,6.1,7.9,7.4]
```

```python
plt.bar(name,score)
plt.show()
```

```python
#Customizing bar graph
plt.figure(figsize=(5,3))
plt.bar(name,score)
plt.title('Scores of Students')
plt.xlabel('Names')
plt.ylabel('Scores')
plt.show()
```

```python
# Changing bar color
plt.figure(figsize=(5,3))
plt.bar(name,score,color='r')
plt.title('Scores of Students')
plt.xlabel('Names')
plt.ylabel('Scores')
plt.show()
```

```python
# Changing bar color using hex codes
# Ref: https://htmlcolorcodes.com/
plt.figure(figsize=(5,3))
plt.bar(name,score,color='#F8ACF1')
plt.title('Scores of Students')
plt.xlabel('Names')
plt.ylabel('Scores')
plt.show()
```

```python
# Generating horizontal bar chart
plt.figure(figsize=(5,3))
plt.barh(name,score,color='#F8ACF1')
plt.title('Scores of Students')
plt.xlabel('Scores')
plt.ylabel('Names')
plt.show()
```

```python
# Changing thickness of the bar
plt.figure(figsize=(5,3))
plt.bar(name,score,color='#F8ACF1',width=0.3)
plt.title('Scores of Students')
plt.xlabel('Names')
plt.ylabel('Scores')
plt.show()
```

```python
# Generating pie chart
population = [1755714,1245413]
labels = ['Rural','Urban']
```

```python
plt.pie(population,labels=labels)
plt.show()
```

```python
# Adding own color to pie chart
totPop = [1755714,1245413]
popType = ['Rural','Urban']
myColors = ['#D55EA2','#05EFCB']
plt.pie(totPop,labels=popType,colors=myColors)
plt.show()
```

```python
# Inserting percentage inside the pie chart
totPop = [1755714,1245413]
popType = ['Rural','Urban']
myColors = ['#D55EA2','#05EFCB']
plt.pie(totPop,labels=popType,colors=myColors,autopct='%.2f%%')
plt.show()
```

```python
# Customizing start angle
totPop = [1755714,1245413]
popType = ['Rural','Urban']
myColors = ['#D55EA2','#05EFCB']
plt.pie(totPop,labels=popType,colors=myColors,autopct='%.2f%%',startangle=90)
plt.show()
```
